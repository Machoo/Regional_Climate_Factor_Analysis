```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("epi_cleaned.csv")
# Only normalize actual features, excluding code, iso, and country
df_normalized = df[df.columns[3:]].copy()
# Normalize
df_normalized = (df_normalized - df_normalized.min()) / (df_normalized.max() - df_normalized.min())
# "Melt" the dataframe to turn features into a "variable"
df_melted = df_normalized.melt(var_name = "Feature", value_name = "Normalized Value")

# Show normalized features to observe general trends, such as whether
# "change over time" variables tend to be high
plt.figure(figsize = (15, 8))
sns.violinplot(x = "Feature", y = "Normalized Value", data = df_melted, inner = "quartile")
plt.xticks(rotation = 90)
plt.xlabel("Feature (TLA)")
plt.title("Violin Plots of Normalized Feature Distributions")
plt.show()
```
